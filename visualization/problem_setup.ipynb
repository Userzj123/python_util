```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
domain = [2*np.pi, np.pi, 1]
dims = [256, 256, 128]


# Define the range and number of points in each direction
x_range = (0, domain[0])
y_range = (0, domain[1])
z_range = (0, domain[2])
range = [x_range, y_range, z_range]
n_x, n_y, n_z = dims

# Generate 1D arrays of x, y, and z coordinates
x_coords = np.linspace(x_range[0], x_range[1], n_x)
y_coords = np.linspace(y_range[0], y_range[1], n_y)
z_coords = np.linspace(z_range[0], z_range[1], n_z)
```

```python
fig =  plt.figure(figsize=(16, 8))
ax = fig.add_subplot(projection='3d')

ax.set_box_aspect(aspect = (x_range[1], y_range[1],z_range[1]))

# Plot the outline of the domain box
for ind, coord in enumerate([x_coords, y_coords, z_coords]):
    tmp = np.zeros(3)
    tmp[ind] += 1
    ax.plot(coord*tmp[0] - range[0][0] * (tmp[0]-1), coord*tmp[1]- range[1][0] * (tmp[1]-1), coord*tmp[2]- range[2][0] * (tmp[2]-1), color='black')
    ax.plot(coord*tmp[0] - range[0][1] * (tmp[0]-1), coord*tmp[1]- range[1][1] * (tmp[1]-1), coord*tmp[2]- range[2][1] * (tmp[2]-1), color='black')
    ax.plot(coord*tmp[0] - range[0][0] * (tmp[0]-1), coord*tmp[1]- range[1][1] * (tmp[1]-1), coord*tmp[2]- range[2][1] * (tmp[2]-1), color='black')
    ax.plot(coord*tmp[0] - range[0][1] * (tmp[0]-1), coord*tmp[1]- range[1][0] * (tmp[1]-1), coord*tmp[2]- range[2][1] * (tmp[2]-1), color='black')
    ax.plot(coord*tmp[0] - range[0][1] * (tmp[0]-1), coord*tmp[1]- range[1][1] * (tmp[1]-1), coord*tmp[2]- range[2][0] * (tmp[2]-1), color='black')




# Setup the axis
ax.set_xlim(x_range[0], x_range[1])
ax.set_ylim(y_range[0], y_range[1])
ax.set_zlim(z_range[0], z_range[1])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
 
ax.grid(visible=None)
# ax.axis('off')

ax.view_init(elev=20, azim=-80, roll=5)
```
