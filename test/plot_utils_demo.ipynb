# Plots related to channel flow Post-processing

## Cross Section contours
Function `contour_channel` plot the cross-sections in xy, xz and yz surfaces at the maximum value given data.

```python
xx = np.linspace(0, 2*np.pi, 100)
yy = np.linspace(0, np.pi, 100)
zz = np.linspace(0, 1, 50)
coords = (xx, yy, zz)

x, y, z = np.meshgrid(xx, yy, zz)
data = np.sin(x)*np.sin(y)*np.sin(z)

fig, ax = plt_utils.contour_channel(coords, data)
```
