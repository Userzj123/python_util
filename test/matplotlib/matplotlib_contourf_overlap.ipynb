```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm 
import scipy.stats as st
import matplotlib

X = np.linspace(0, 2*np.pi, 128)
Y = np.linspace(0, 2*np.pi, 128)

r = lambda x, y: x[:, np.newaxis] + y[np.newaxis, :]

g = lambda x, y: np.sin(x)[:, np.newaxis]*np.sin(y)[np.newaxis, :]

b = lambda x, y: np.sinh(x)[:, np.newaxis]*np.sinh(y)[np.newaxis, :]


CSr = plt.contourf(X,Y, r(X, Y), levels=50,
                     cmap=cm.Reds, )

plt.colorbar()
```

```python
CSg = plt.contourf(X,Y,g(X, Y), levels=50,
                     cmap=cm.Greens, )

plt.colorbar()
```

```python
CSb = plt.contourf(X,Y,b(X, Y), levels=50,
                     cmap=cm.Blues)

plt.colorbar()
```

```python
z = np.zeros(shape=(128, 128, 4))
z[:, :, 0] = r(X, Y)
z[:, :, 1] = g(X, Y)
z[:, :, 2] = b(X, Y)
z[:, :, 3] = 0.5

plt.imshow(z)
```
